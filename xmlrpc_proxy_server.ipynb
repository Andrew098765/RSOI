```python
# xmlrpc_server.ipynb

from xmlrpc.server import SimpleXMLRPCServer
from xmlrpc.server import SimpleXMLRPCRequestHandler
import xmlrpc.client
import datetime
import pandas as pd
import pickle

class RequestHandler(SimpleXMLRPCRequestHandler):
    rpc_paths = ('/RPC2',)

proxy_server = SimpleXMLRPCServer(("localhost", 8009), 
                            requestHandler=RequestHandler)

#server.register_introspection_functions()

#stats_server = xmlrpc.client.ServerProxy("http://localhost:8018")
stats_server = xmlrpc.client.ServerProxy("http://localhost:7000")
server = xmlrpc.client.ServerProxy("http://localhost:8008")


def get_event_logs(operation_type, start_time, end_time):
    stat = stats_server.get_event_log(operation_type, start_time, end_time)
    return stat

proxy_server.register_function(get_event_logs, 'get_event_logs')

# Добавление в лог через сервер
def add_log(types, time, delta):
    print('test1')
    try:
        stats_server.save_event(types, time, delta)
    finally:
        
        return True
    
# Тест
def ping():
    print('test2')
    time = datetime.datetime.now()
    result = server.ping()
    delta = datetime.datetime.now() - time
    add_log("ping", time, delta)
    return ping()
proxy_server.register_function(ping, 'ping')

# Время сервера
def now():
    print('test3')
    time = datetime.datetime.now()
    result = server.now()
    delta = datetime.datetime.now() - time
    add_log("now", time, delta)
    return now()

# Отображение строкового вида, типа и значений
def show_type(arg):
    print('test4')
    time = datetime.datetime.now()
    result = server.show_type()
    delta = datetime.datetime.now() - time
    add_log("type", time, delta)
    return show_type()
proxy_server.register_function(show_type, 'type')

# Сумма
def test_sum(a, b):
    print('test5')
    time = datetime.datetime.now()
    result = server.test_sum()
    delta = datetime.now() - time
    add_log("sum", time, delta)
    return test_sum()
proxy_server.register_function(test_sum, 'sum')

# Степень
def test_pow(a, b):
    time = datetime.datetime.now()
    result = server.test_pow()
    delta = datetime.datetime.now() - time
    add_log("pow", time, delta)
    return test_pow()
proxy_server.register_function(test_pow, 'pow')

# Проверка нахождения клиента в черном списке c использованием Pandas Data Frame
def black_list_check(sname):
    time = datetime.datetime.now()
    result = server.black_list_check()
    delta = datetime.datetime.now() - time
    add_log("black_list_check", time, delta)
    return black_list_check()

proxy_server.register_function(black_list_check, 'black_list_check')

# Проверка нахождения клиента в черном списке c использованием Pandas Data Frame
def black_list_check_par(sname, name, patronym, date):
    time = datetime.datetime.now()
    result = server.black_list_check_par()
    delta = datetime.datetime.now() - time
    add_log("black_list_check_par", time, delta)
    return black_list_check_par()

proxy_server.register_function(black_list_check_par, 'black_list_check_par')

# Бинарная передача данных
def send_back_binary(bin_data):
    time = datetime.datetime.now()
    result = server.send_back_binary()
    delta = datetime.datetime.now() - time
    add_log("send_back_binary", time, delta)
    return send_back_binary()
proxy_server.register_function(send_back_binary, 'send_back_binary')

# бинаризация с параметром
def send_back_binary_par(bin_data, threshold):
    time = datetime.datetime.now()
    result = server.send_back_binary_par()
    delta = datetime.datetime.now() - time
    add_log("send_back_binary_par", time, delta)
    return send_back_binary_par()
proxy_server.register_function(send_back_binary_par, 'send_back_binary_par')

# Инверсия цвета
# На вход изображение RGB размерности (M, N, 3) со значениями 0-255
def send_back_inversion(bin_data):
    time = datetime.datetime.now()
    result = server.send_back_inversion()
    delta = datetime.datetime.now() - time
    add_log("send_back_inversion", time, delta)
    return send_back_inversion()
proxy_server.register_function(send_back_inversion, 'color_inversion')

# перевенуть картинку в вертикали
def send_back_flipped(bin_data):
    time = datetime.datetime.now()
    result = server.send_back_flipped()
    delta = datetime.datetime.now() - time
    add_log("send_back_flipped", time, delta)
    return send_back_flipped()
proxy_server.register_function(send_back_flipped, 'send_back_flipped')

print("Listening on port 8009...")
proxy_server.serve_forever()

```
